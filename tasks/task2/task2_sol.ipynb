```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("./cartoon1.pgm", cv2.IMREAD_GRAYSCALE)
imgs = []
sigma = 1.6

plt.figure(figsize=(15, 8))
k = 1.15
for i in range(5):
    imgs.append(img)

    s = sigma * k**i
    img = cv2.GaussianBlur(img, (17, 17), s, s)

    # Plot the blurred images (2 points)
    plt.subplot(1, 5, i + 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), cmap="gray")
    plt.title(f"Blur sigma={s}")

plt.show()
```

```python
DoGs = []

plt.figure(figsize=(15, 8))

# correct DOG calculation (2 points)
for i in range(4):
    DoG = imgs[i + 1] - imgs[i]
    DoGs.append(DoG)

    # Plot the blurred image (3 points)
    plt.subplot(1, 4, i + 1)
    plt.imshow(cv2.cvtColor(DoG, cv2.COLOR_BGR2RGB), cmap="gray")
    plt.title(f"DoG {i + 1}")

plt.show()
```

```python
img = cv2.imread("./cartoon1.pgm", cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, (0, 0), fx=0.5, fy=0.5)
imgs = []
sigma = 1.6
k = 1.15

plt.figure(figsize=(15, 8))

for i in range(5):
    imgs.append(img)

    s = sigma * k**i
    img = cv2.GaussianBlur(img, (17, 17), s, s)

    # Plot the blurred image
    plt.subplot(1, 5, i + 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), cmap="gray")

    plt.title(f"Blur {i}")

plt.show()

DoGs = []

plt.figure(figsize=(15, 8))

for i in range(4):
    DoG = imgs[i + 1] - imgs[i]
    DoGs.append(DoG)

    # Plot the blurred image
    plt.subplot(1, 4, i + 1)
    plt.imshow(cv2.cvtColor(DoG, cv2.COLOR_BGR2RGB), cmap="gray")
    plt.title(f"DoG {i + 1}")

# 1 point
plt.show()
```

```python
# SIFT
image = cv2.imread("./cartoon1.pgm", cv2.IMREAD_GRAYSCALE)
image = cv2.GaussianBlur(image, (17, 17), 1.6, 1.6)

sift = cv2.SIFT_create(nfeatures=250)
kp = sift.detect(image)

plt.figure(figsize=(10, 10))
img = cv2.drawKeypoints(image, kp, -1, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img)
```
